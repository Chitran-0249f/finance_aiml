## Step 2: Load Preprocessed Data from Parquet

```python
# Define paths using pathlib (production-standard approach)
PROJECT_ROOT = Path('..')
DATA_DIR = PROJECT_ROOT / 'data'
DF_EDA_PATH = DATA_DIR / 'df_eda_features.parquet'

# Load the parquet file
df = pd.read_parquet(DF_EDA_PATH)

print(f"✅ Loaded data from {DF_EDA_PATH}")
print(f"Shape: {df.shape}")
print(f"\nColumns: {df.columns.tolist()}")
print(f"\nData types:\n{df.dtypes}")
```

## Step 6: Save Engineered Features

```python
# Define output path
FEATURES_OUTPUT_PATH = DATA_DIR / 'df_features_engineered.parquet'

# Save the engineered features
df_features.to_parquet(FEATURES_OUTPUT_PATH, index=False)

print(f"✅ Saved engineered features to {FEATURES_OUTPUT_PATH}")
print(f"File size: {FEATURES_OUTPUT_PATH.stat().st_size / (1024**2):.2f} MB")
```
